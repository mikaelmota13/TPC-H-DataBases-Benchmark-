```python
df = pd.read_csv("means_equalized.csv", index_col=0)

x_labels = df.index.astype(str).tolist()
y = pd.to_numeric(df["mean_us"], errors="coerce").values
x = range(len(x_labels))

fig, ax = plt.subplots()
bars = ax.bar(x, y)

ax.set_xticks(x)
ax.set_xticklabels(x_labels, rotation=90)
ax.set_xlabel("Query")
ax.set_ylabel("mean_us")
ax.yaxis.set_major_formatter(StrMethodFormatter('{x:,.0f}'))  # eixo Y sem decimais

# rÃ³tulos inteiros no topo das barras
ax.bar_label(bars, labels=[f"{int(round(v))}" for v in y], padding=2, fontsize=8)

plt.tight_layout()
plt.show()
```
